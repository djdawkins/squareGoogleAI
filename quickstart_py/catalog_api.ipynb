```python
from square.client import Client
import os
from dotenv import load_dotenv
import pandas as pd
load_dotenv()
```

```python
client = Client(
    access_token=os.environ['SQUARE_ACCESS_TOKEN'],
    environment='sandbox')

orders_api = client.orders
```

```python
# Search orders
body = {
    "object_types": [
        "ITEM"
    ],
    "query": {
        "sorted_attribute_query": {
          "attribute_name": "name"
        }
    },
    "limit": 100
}
result = client.catalog.search_catalog_objects(body)

if result.is_success():
  print(result.body)
elif result.is_error():
  print(result.errors)
```

```python
catalog_dict = result.body
catalog_objects_list = catalog_dict["objects"]

order_entries_list = catalog_objects_list

print(len(catalog_objects_list))
cat_id_list = []

for x in range(0, len(catalog_objects_list)):
    cat_id = catalog_objects_list[x]["id"]
    cat_id_list.append(cat_id)

cat_id_list
```

```python
batch_catalogs = client.catalog.batch_retrieve_catalog_objects(
  body = {
    "object_ids": cat_id_list,
    "include_related_objects": True
  }
)

if batch_catalogs.is_success():
  print(batch_catalogs.body)
elif batch_catalogs.is_error():
  print(batch_catalogs.errors)
```

```python
batch_catalogs_dict = batch_catalogs.body
batch_catalogs_list = batch_catalogs_dict['objects']
batch_catalogs_list
```

```python
cat_list = []
for cat in batch_catalogs_list:
    cat_obj_id = cat["id"]
    item_data_name = cat["item_data"]["name"]
    cat_type = cat["type"]
    updated_at = cat["updated_at"]
    is_deleted = cat["is_deleted"]
    version = cat["version"]
    present_at_all_locations = cat["present_at_all_locations"]
    # customer_id = cat["customer_id"]

    cat_list.append((cat_obj_id, item_data_name, cat_type, updated_at, version, present_at_all_locations))


# create DataFrame using data
cat_df = pd.DataFrame(cat_list, columns =["cat_obj_id", "item_data_name", "cat_type", "updated_at", "version", "present_at_all_locations"])

cat_df.to_csv("./graph_data/catolgs.csv", index=False)
cat_df
```
